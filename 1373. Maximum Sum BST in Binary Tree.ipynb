# [1373. Maximum Sum BST in Binary Tree](https://leetcode.com/problems/maximum-sum-bst-in-binary-tree/)

Given a binary tree root, the task is to return the maximum sum of all keys of any sub-tree which is also a Binary Search Tree (BST).

Assume a BST is defined as follows:

- The left subtree of a node contains only nodes with keys less than the node's key.
- The right subtree of a node contains only nodes with keys greater than the node's key.
- Both the left and right subtrees must also be binary search trees.

## 思路版

题目的目标是求最大的二叉搜索树总和，那么我们可以逐步拆解，从这个角度入手：1.判断一个节点为根时能否构成二叉搜索树 2.求这个二叉搜索树的和！

### 1.先判断是否为二叉搜索树

```
def isBST(node,min_val,max_val):
    if node is None:
        return True
        
    if node.val <= min_val or node.val >= max_val:
        return False
        
    return self.isBST(node.left,min_val,node.val) and self.isBST(node.right,node.val,max_val)

# 调用
isBST(root,float('-inf'),float('inf'))
```

### 2.判断每一个节点为根节点时是否构成二叉树

应该还可以细分两种形式：从上到下或者从下到上！

从上到下版本根据上面的代码，然后再写一次遍历的过程依次判定就行了，实现比较简单但是效率比较低。

所以重点看下从下到上的版本：

```
        res = {0:True}
        def dfs_bst(node):
            if not node:
                return TreeNode(0)
            
            left = dfs_bst(node.left)
            right = dfs_bst(node.right)
            
            if res[left.val] and res[right.val]:
                if left.val == 0 and right.val == 0:
                    res[node.val] = True
                elif left.val == 0 and right.val > node.val:
                    res[node.val] = True
                elif right.val == 0 and left.val < node.val:
                    res[node.val] = True
                elif left.val < node.val < right.val:
                    res[node.val] = True
                else:
                    res[node.val] = False
            else:
                res[node.val] = False
                
            return node
            
        dfs_bst(root)
        print(res)
```
