# [513. Find Bottom Left Tree Value](https://leetcode.com/problems/find-bottom-left-tree-value/)

Given a binary tree, find the leftmost value in the last row of the tree.

## 自己版

看完题目，第一次把题目理解错了，以为是求最深层且一定是左子节点的节点。提交了个答案发现要求的是最后一行的最左的节点。

## 借助列表的长度与层级之间的关系：

```python
    def findBottomLeftValue(self, root: TreeNode) -> int:
        res = []
        def dfs(node,level):
            if node:
                if level == len(res):
                    res.append(node.val)
                    
                dfs(node.left, level+1)
                dfs(node.right, level+1)
                
        dfs(root,0)
        return res[-1]
```

## 上一版的优化 - 将列表替换为一个变量后计算与层级之间的关系：

```python
    def findBottomLeftValue(self, root: TreeNode) -> int:
        
        self.depth = 0
        self.res = 0
        
        def dfs(node, level):
            if node:
                if self.depth == level:
                    self.depth += 1
                    self.res = node.val
                   
                dfs(node.left, level+1)
                dfs(node.right, level+1)
                
        dfs(root, 0)
        return self.res
```

## Right-to-left BFS

```python
    def findBottomLeftValue(self, root: TreeNode) -> int:
        queue = [root]
        for node in queue:
            if node.right:
                queue += [node.right]
                
            if node.left:
                queue += [node.left]
                
        return node.val
```
