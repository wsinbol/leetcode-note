# [652. Find Duplicate Subtrees](https://leetcode.com/problems/find-duplicate-subtrees/)

Given a binary tree, return all duplicate subtrees. For each kind of duplicate subtrees, you only need to return the root node of any one of them.

Two trees are duplicate if they have the same structure with same node values.

## 自己版

抛开返回值的限制，程序是没问题的！但是题目要求返回 List[TreeNode]，故无法AC!

```python
    def findDuplicateSubtrees(self, root: TreeNode) -> List[TreeNode]:
        res = {}
        def dfs(node):
            if not node:
                return ''
            left = dfs(node.left)
            right = dfs(node.right)
            c = str(node.val)
            if left:
                c += ',' + left
            if right:
                c += ',' + right
            res.setdefault(c,0)
            res[c] += 1
            return c
            
        dfs(root)
        ret = []
        for k,v in res.items():
            if v > 1:
                ret.append([TreeNode(i) for i in k.split(',')])
        return ret
```

## Bug版

```python
from collections import defaultdict
class Solution:
    def findDuplicateSubtrees(self, root: TreeNode) -> List[TreeNode]:
        res = defaultdict(list)
        def dfs(node):
            if not node:
                return ''
            left = dfs(node.left)
            right = dfs(node.right)
            c = str(node.val)
            if left and left != c:
                c += ',' + left
            if right and right != c:
                c += ',' + right
            res[c].append(node)
            return c
            
        dfs(root)
        ret = []
        for k,v in res.items():
            if len(v) > 1:
                ret.append(v[0])
        return ret
```

## AC版

```python
from collections import defaultdict
class Solution:
    def findDuplicateSubtrees(self, root: TreeNode) -> List[TreeNode]:
        res = defaultdict(list)
        def dfs(node):
            if not node:
                return 'null'
            left = dfs(node.left)
            right = dfs(node.right)
            c = str(node.val)
            if left and left != c:
                c += ',' + left
            if right and right != c:
                c += ',' + right
            res[c].append(node)
            return c
            
        dfs(root)
        ret = []
        for k,v in res.items():
            if len(v) > 1:
                ret.append(v[0])
        return ret
```

# 总结

bug版 和 AC版 的不同之处在于节点为空时的返回值,前者返回 空，而后者返回 字符串'null'，尤其是下面的两个例子格外需要注意：

```
1. [0,0,0,0,null,null,0,null,null,0] # return [[0],[0,0]]
2. [0,0,0,0,null,null,0,null,null,null,0] # return [[0]]
```

也就是说 涉及到将树的遍历转成字符串作为字典的 key 值时，一定要将空节点的返回值设置为 'null'！
