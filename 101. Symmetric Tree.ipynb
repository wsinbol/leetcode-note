# [101. Symmetric Tree](https://leetcode.com/problems/symmetric-tree/)

Given a binary tree, check whether it is a mirror of itself (ie, symmetric around its center).

For example, this binary tree [1,2,2,3,4,4,3] is symmetric:

```
    1
   / \
  2   2
 / \ / \
3  4 4  3
```

But the following [1,2,2,null,3,null,3] is not:

```
    1
   / \
  2   2
   \   \
   3    3
```

## 思路版：

题目中要判断二叉树的左子树和右子树是否关于根节点对称，想了一下：左子树采用中序遍历，而右子树采用反向中序遍历，对比两次遍历的结果，如果相等则证明二者是堆成的！又仔细想了一下：这么判断是不全面的，在 Example 2 的基础上稍微修改一下：[1,2,3,null,3,null,2]，左右遍历的结果相同，但是这棵树明显不是堆成的二叉树。所以此方法有待改进：加入层级怎么样呢？

```
    def isSymmetric(self, root: TreeNode) -> bool:
        left,right = [],[]
        def dfs_left(node):
            if node:
                dfs_left(node.left,)
                left.append(node.val)
                dfs_left(node.right)
                
        def dfs_right(node):
            if node:
                dfs_right(node.right)
                right.append(node.val)
                dfs_right(node.left)
                
        dfs_left(root.left)
        dfs_right(root.right)
        print(left)
        print(right)
```

### 加入层级：

```
    def isSymmetric(self, root: TreeNode) -> bool:
        if not root:
            return True
        
        left,right = [],[]
        def dfs_left(node,level):
            if node:
                if node.left:
                    dfs_left(node.left,level+1)
                left.append((node.val,level))
                if node.right:
                    dfs_left(node.right,level+1)
                
        def dfs_right(node,level):
            if node:
                if node.right:
                    dfs_right(node.right,level+1)
                right.append((node.val, level))
                if node.left:
                    dfs_right(node.left,level+1)
                
        dfs_left(root.left,0)
        dfs_right(root.right,0)
        return left == right
```

## 官方版

DFS版：

```
    def isSymmetric(self, root: TreeNode) -> bool:
        def dfs(node1, node2):
            if node1 is None and node2 is None:
                return True
            
            if node1 is None or node2 is None:
                return False
            
            return (node1.val == node2.val) and dfs(node1.left, node2.right) and dfs(node1.right, node2.left)
        
        return dfs(root,root)
```

BFS版：

```
    def isSymmetric(self, root: TreeNode) -> bool:
        queue = [root, root]
        while len(queue) > 0:
            node1 = queue.pop()
            node2 = queue.pop()
            
            if node1 is None and node2 is None:
                continue
            if node1 is None or node2 is None:
                return False
            if node1.val != node2.val:
                return False
            
            queue.append(node1.left)
            queue.append(node2.right)
            queue.append(node1.right)
            queue.append(node2.left)
            
        return True
```
