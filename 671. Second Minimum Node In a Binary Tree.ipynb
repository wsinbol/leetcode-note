# [671. Second Minimum Node In a Binary Tree](https://leetcode.com/problems/second-minimum-node-in-a-binary-tree/)

Given a non-empty special binary tree consisting of nodes with the non-negative value, where each node in this tree has exactly two or zero sub-node. If the node has two sub-nodes, then this node's value is the smaller value among its two sub-nodes. More formally, the property root.val = min(root.left.val, root.right.val) always holds.

Given such a binary tree, you need to output the second minimum value in the set made of all the nodes' value in the whole tree.

If no such second minimum value exists, output -1 instead.

## 自己版

以BFS为核心的遍历查找实现方法，这里需要注意一下 tmp 的赋值，是取 node.left 和 node.right 中的最大值。

```
    def findSecondMinimumValue(self, root: TreeNode) -> int:
        if not root:
            return -1
        
        queue = [root]
        res = [root.val,float('inf')]
        
        while len(queue) > 0:
            size = len(queue)
            for i in range(size):
                current = queue.pop(0)
                
                if current.left and current.right:
                
                    # Attention
                    tmp = max(current.left.val, current.right.val)
                    if tmp < res[1] and tmp > res[0]:
                        res[1] = tmp
                    
                if current.left:
                    queue.append(current.left)
                if current.right:
                    queue.append(current.right)
                
        return res[1] if res[1] < float('inf') else -1
```

## 自我衍生版

理论上说，DFS 和 BFS 的时间复杂度应该是差不多的，都是要遍历完这课树所有的节点，如果执意想降低 时间复杂度，根据题目的描述和测试用例的特点，即 root 节点的值是最小的，我们可以预判如果当遍历的节点为 root.val + 1 时，遍可以提前终止程序，这个样子肯定可以大幅提高极端数据的情况了！

```
def findSecondMinimumValue(self, root: TreeNode) -> int:

    if not root:
        return -1

    res = float('inf')
    queue = [root]

    while len(queue) > 0:
        size = len(queue)
        for i in range(size):
            current = queue.pop(0)

            if current and root.val < current.val < res:
                res = current.val
                # 提前终止
                if res == root.val + 1:
                    return res

            if current.left:
                queue.append(current.left)

            if current.right:
                queue.append(current.right)

    return -1 if res == float('inf') else res
```

## DFS 版

```
def findSecondMinimumValue(self, root: TreeNode) -> int:

    self.res = float('inf')

    def dfs(node):
        if not node:
            return

        if root.val < node.val < self.res:
            self.res = node.val

        dfs(node.left)
        dfs(node.right)

    dfs(root)
    return -1 if self.res == float('inf') else self.res
```
