# [1367. Linked List in Binary Tree](https://leetcode.com/problems/linked-list-in-binary-tree/)

Given a binary tree root and a linked list with head as the first node. 

Return True if all the elements in the linked list starting from the head correspond to some downward path connected in the binary tree otherwise return False.

In this context downward path means a path that starts at some node and goes downwards.

## DIY

思路还是没问题的，但是超时了。。。

```python
class Solution:
    def isSubPath(self, head: ListNode, root: TreeNode) -> bool:
        target = []
        while head:
            target.append(head.val)
            head = head.next
            
        def dfs(node, path):
            if node:
                # print(path+[node.val])
                # if path+[node.val] == target or (len(target) == 1 and node.val in target):
                if path+[node.val] == target or path == target:
                    return True
                    
                return dfs(node.left, path+[node.val]) or dfs(node.left, [node.val]) or dfs(node.right, path+[node.val]) or dfs(node.right, [node.val])
                
                
        return dfs(root, [])
```

## Upgrade

give a example:
```python
a = [1,2,3,4]
b = [1,2]
b in a # always return False
```

So the solution above is wrong!

```python
class Solution:
    def isSubPath(self, head: ListNode, root: TreeNode) -> bool:
        target = ''
        while head:
            target += str(head.val) + ','
            head = head.next
        
        def dfs(node, path):
            if target in path:
                return True
            if node:
                return dfs(node.left, path+str(node.val)+',') or dfs(node.right, path+str(node.val)+',')
                
        return dfs(root, '')
```

```python
class Solution:
    def isSubPath(self, head: ListNode, root: TreeNode) -> bool:
        DELIMITER = '|'
        subpath = ''
        while head:
            subpath += str(head.val) + DELIMITER
            head = head.next
    
        def dfs(root, path):
            if subpath in path:
                return True
            if not root:
                return False
            return dfs(root.left, path+str(root.val)+DELIMITER) or dfs(root.right, path+str(root.val)+DELIMITER)
            
        return dfs(root, "")
```

## Niubility

写的非常巧妙！

```python
class Solution:
    def isSubPath(self, head: ListNode, root: TreeNode) -> bool:
        if not head:
            return True
        if not root:
            return False
        
        def dfs(head, root):
            if not head:
                return True
            if not root:
                return False
            return head.val == root.val and (dfs(head.next, root.left) or dfs(head.next, root.right))
        
        return dfs(head, root) or self.isSubPath(head, root.left) or self.isSubPath(head, root.right)
```
