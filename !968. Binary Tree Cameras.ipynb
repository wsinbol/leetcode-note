# [968. Binary Tree Cameras](https://leetcode.com/problems/binary-tree-cameras/)

Given a binary tree, we install cameras on the nodes of the tree. 

Each camera at a node can monitor its parent, itself, and its immediate children.

Calculate the minimum number of cameras needed to monitor all nodes of the tree.

## 思考版

这道题的难度在于对状态的界定！

```
#0：该节点有监控器，可以被监控
#1：该节点无监控器，但可被监控
#2：该节点无监控器，不可被监控
```

伪代码描述一下状态：

```

dfs(node) 返回的是当前 node 节点的状态，在写 递归的过程中，千万不要想复杂，只按照最简单的实例来描述这个过程即可。假设目前只有三个节点的二叉树：

left = dfs(node.left) 表示的是 当前节点的 左子节点的监控状态
right = dfs(node.right) 表示的是 当前节点的 右子节点的监控状态

1.if left == 0 or right == 0:表示 左子节点 或者 右子节点 有监控器，而当前节点又是父节点，故整个 dfs(node) 应该返回 #1
2.if right == 2 or left == 2:表示 左子节点 或者 右子节点 既没有监控器也没有处在监控状态，此时应该给当前节点添加一个监控器，然后返回 #0
3.if left == 1 or right == 1:表示 左子节点 或者 右子节点 无监控器，但可处于被监控状态，那么此时的情况又复杂了一点：第一种情况是父节点（即当前节点）有一个监控器，这样对于当前节点来说应该是和 2 重复了，添加一个监控器后返回 #0(和 2 冲突了！故不可取！)；第二种情况是直接返回 #2，然后在外面判断进行 +1 的操作！

```

下面来看完整的代码：

```
    def minCameraCover(self, root: TreeNode) -> int:
        if not root:
            return 0
        self.res = 0
        def dfs(node):
            if node == None:
                return 1
            
            left = dfs(node.left)
            right = dfs(node.right)
            
            if left == 2 or right == 2:
                self.res += 1
                return 0
            elif left == 0 or right == 0:
                return 1
            else:
                return 2
        
        if dfs(root) == 2:
            self.res += 1
        return self.res
```
