# 1.Depth of Binary Tree

分别求二叉树的最大深度和最小深度，BFS、递归方法都可以解决二者的问题。

## BFS方法

先看下BFS实现的二者的代码区别，尤其注意 depth 的值：

```
#1 最大深度代码
def maxDepth(self, root: TreeNode) -> int:
    if not root:
        return 0

    queue = [root]
    depth = 0 # 注意这里
    while len(queue) > 0:
        size = len(queue)
        for i in range(size):
            current = queue.pop(0)
            if current.left:
                queue.append(current.left)
            if current.right:
                queue.append(current.right)
        # 加1操作的位置可以放在 for...in...前面
        depth += 1

    return depth
```

```
#2 最小深度代码
def minDepth(self, root: TreeNode) -> int:
    if not root:
        return 0

    queue = [root]
    depth = 1 # 注意这里

    while len(queue) > 0:
        size = len(queue)
        for i in range(size):
            current = queue.pop(0)
            # 注意这里的逻辑是 and !!!
            if current.left == None and current.right == None:
                return depth

            if current.left:
                queue.append(current.left)
            if current.right:
                queue.append(current.right)
        # 加1操作的位置不能放在前面
        depth += 1

    return depth
```

<font color=red>
    问题：同样都是先将 root 加入到 queue 中，为啥求最大深度时 depth 从 0 开始，而求最小深度时 depth 从 1 开始？<br>
    理解：求最大深度时 depth 不会在循环体内中途返回，而且最终的 depth 会在循环体内 +1 进行弥补；而求最小深度时，depth 在循环体内会中途返回结果，故 depth +1 操作不会执行到，这也看出 depth += 1 不可以放在 for...in 结构前面，这样比预期的结果多1！
</font>

## 递归法

```
def maxDepth(self, root: TreeNode) -> int:
    if not root:
        return 0

    left = self.maxDepth(root.left)
    right = self.maxDepth(root.right)

    return max(left, right) + 1
```

```
def minDepth(self, root: TreeNode) -> int:
    if not root:
        return 0

    left = self.minDepth(root.left)
    right = self.minDepth(root.right)

    return left+right+1 if left == 0 or right == 0 else min(left,right) + 1
```

## 二者的优劣

```
BFS is much better here, rather than a DFS approach.

Sure, the solution here is short in terms of lines of code and looks nice, but it's far from optimal.

If you have a tree where say the root's left subtree has a depth of 500 and the right subtree has a depth of 1, the code is going to traverse all the way down the 500 left subtree first before finally traversing the right subtree with a measly depth of 1 and figuring out "d'oh!" that's the min depth.

With BFS, instead of traversing 501 nodes to figure out the min depth, you could've just traversed two. Now imagine if the left subtree comprised of tens of thousands of nodes ...
```

# 2.Height of Binary Tree

## DFS方法

```
# 求二叉树节点的最大高度
def getHeight(node):
    if not node:
        return -1

    left = getHeight(node.left)
    right = getHeight(node.right)
    return max(left, right) + 1
```

```
# 求二叉树节点的深度
def maxDepth(self, root: TreeNode) -> int:
    if not root:
        return 0

    left = self.maxDepth(root.left)
    right = self.maxDepth(root.right)

    return max(left, right) + 1
```

<font color=red>
对于深度而言，node 节点为空时，返回 0;<br>
对于高度而言，node 节点为空时，返回 -1（即 root 节点也是空的，该树是一个空树）;<br>
其实，求高度时也可以返回 0，但是需要提前返回值！
</font>
