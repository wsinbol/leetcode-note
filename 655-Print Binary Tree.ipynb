# 655-Print Binary Tree

[655 Print Binary Tree](https://leetcode.com/problems/univalued-binary-tree/)

## 分析

题目描述可以发现二维数组的行数：树的高度，所以需要求整个树的高度；而二维数组的列数为 (2**树的高度)-1；故可以先用""填充整个二维数组。

接下来就是将树中每个node的值填充到对应的二维数组中，此时需要发现规律，以 Example 2 为例：

row count:3 

column count: 2**3-1=7

| node | row_i | column_i | remark | rule |
| ----------- | ----------- | ----------- | ----------- | ----------- |
| 1 | 0 | 3 | (0+7)//2=3 |
| 2,3 | 1 | 1,5 | (0+3)//2=1,(4+7)//2=5 |
| 4 | 2 | 2 | (0+1)//2=0(2的左节点)，(2+3)//2=2(2的右节点) |

## 递归版

```
def printTree(self, root: TreeNode) -> List[List[str]]:

    def getHeight(node):
        if not node:
            return 0

        left = getHeight(node.left)
        right = getHeight(node.right)
        return max(left,right)+1

    m = getHeight(root)
    n = 2**m-1

    draw = [["" for _ in range(n)] for _ in range(m)]
    
    # 1
    # space 表示一个半区的长度
    def fill(node,mi,ni,space):
        if node:
            draw[mi][ni] = str(node.val)
            fill(node.left,mi+1,ni-space, space//2)
            fill(node.right,mi+1,ni+space, space//2)
    fill(root,0,n//2,(n+1)//4)
    
    # 2
    def fill(node,i,l,r):
        if node:
            draw[i][(l+r)//2] = str(node.val)
            fill(node.left,i+1, l, (l+r)//2)
            fill(node.right,i+1, r, (l+r+1)//2)
    
    return draw
```

## 迭代版

迭代版的思路其实就是将递归的参数传递计算过程转成字典的形式存储起来，然后再遍历树的结点时根据预先计算好的字典计算对应的位置即可。一般都要辅助队列或者栈来实现。
