# [!116&117. Populating Next Right Pointers in Each Node](https://leetcode.com/problems/populating-next-right-pointers-in-each-node/)

## 116

You are given a perfect binary tree where all leaves are on the same level, and every parent has two children. The binary tree has the following definition

```c
struct Node {
  int val;
  Node *left;
  Node *right;
  Node *next;
}
```

Populate each next pointer to point to its next right node. If there is no next right node, the next pointer should be set to NULL.

Initially, all next pointers are set to NULL.

```
Follow up:

You may only use constant extra space.
Recursive approach is fine, you may assume implicit stack space does not count as extra space for this problem.
```

```python

def connect(self, root: 'Node') -> 'Node':
    stack = []
    def dfs(node,pos,level):
        if node:
            if len(stack) == level:
                stack.append([])
            stack[level].append(node)
            left = dfs(node.left, pos*2,level+1)
            right = dfs(node.right, pos*2+1,level+1)
            return node

    dfs(root,1,0)
    for i in stack:
        j = 0
        while j < len(i):
            n = j + 1
            if n < len(i):
                i[j].next = i[n]
            else:
                i[j].next = None

            j += 1

    return root
```

## 借助哨兵简化处理方式

```python
    def connect(self, root: 'Node') -> 'Node':
        curr = root # 当前节点，开始准备处理其子节点
        prev = None # 下一层要处理的节点的前缀
        head = None # 下一层要处理的头结点
        
        while curr != None:
            while curr != None:
                if curr.left != None:
                    if prev != None:
                        prev.next = curr.left
                    else:
                        head = curr.left
                    prev = curr.left
                    
                if curr.right != None:
                    if prev != None:
                        prev.next = curr.right
                    else:
                        head = curr.right
                    prev = curr.right
                    
                curr = curr.next
                
            curr = head
            head = None
            prev = None
            
        return root

```
