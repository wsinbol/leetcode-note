# [1339. Maximum Product of Splitted Binary Tree](https://leetcode.com/problems/maximum-product-of-splitted-binary-tree/)

Given a binary tree root. Split the binary tree into two subtrees by removing 1 edge such that the product of the sums of the subtrees are maximized.

Since the answer may be too large, return it modulo 10^9 + 7.

## 自己版

从一棵二叉树中找到两个根节点，其中一个根节点是原来的根节点，另外一个根节点是原根节点的子节点，然后分别计算这两个根节点的和，并将两个和相乘就是最终的结果。我写的如下：

```
    def maxProduct(self, root: TreeNode) -> int:
        
        def getSum(node,new_node):
            
            if not node:
                return 0
            
            left = 0
            right = 0
            
            if node.left != new_node:
                left = getSum(node.left,new_node)
            if node.right != new_node:
                right = getSum(node.right,new_node)
            
            return node.val + left + right
                
        def splitTree(root_node):
            queue = [root_node]
            res = []
            while len(queue) > 0:
                size = len(queue)
                for i in range(size):
                    current = queue.pop(0)
                    
                    if current.left:
                        p = getSum(root_node, current.left)
                        q = getSum(current.left, None)
                        res.append(p*q)
                        queue.append(current.left)
                        
                    if current.right:
                        p = getSum(root_node, current.right)
                        q = getSum(current.right, None)
                        res.append(p*q)
                        queue.append(current.right)
            return res
        
        return max(splitTree(root)) % (10**9+7)
```

```
    def maxProduct(self, root: TreeNode) -> int:
        
        def getSum(node,new_node):
            
            if not node:
                return 0
            
            left = 0
            right = 0
            
            if node.left != new_node:
                left = getSum(node.left,new_node)
            if node.right != new_node:
                right = getSum(node.right,new_node)
            
            return node.val + left + right
                
        def splitTree(root_node):
            queue = [root_node]
            res = []
            total = getSum(root_node, None)
            while len(queue) > 0:
                size = len(queue)
                for i in range(size):
                    current = queue.pop(0)
                    
                    if current.left:
                        p = getSum(root_node, current.left)
                        q = total - p
                        res.append(p*q)
                        queue.append(current.left)
                        
                    if current.right:
                        p = getSum(root_node, current.right)
                        q = total - p
                        res.append(p*q)
                        queue.append(current.right)
            return res
        
        return max(splitTree(root)) % (10**9+7)
```

## DFS版

后续遍历：

```
    def maxProduct(self, root: TreeNode) -> int:
        sums = []
        def dfs(node):
            if not node:
                return 0
            res = node.val + dfs(node.left) + dfs(node.right)
            sums.append(res)
            return res
        
        total = dfs(root)
        return max([i*(total-i) for i in sums]) % (10**9+7)
```
