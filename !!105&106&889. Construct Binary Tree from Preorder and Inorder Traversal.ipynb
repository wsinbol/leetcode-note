## 105 Preorder and Inorder

### Recursive Solution

前序前找根，中序分左右<br>
前序直接传，中序索引割

#### 1. easy to understand but low efficiency

```python
    def buildTree(self, preorder: List[int], inorder: List[int]) -> TreeNode:
        if inorder:
            current = preorder.pop(0)
            index = inorder.index(current)
            node = TreeNode(current)
            node.left = self.buildTree(preorder,inorder[:index])
            node.right = self.buildTree(preorder, inorder[index+1:])
            return node
```

#### 2. imporve efficiency

index = inorder.index(current) 的性能不如 id_map = {val:idx for idx,val in enumerate(inorder)}！

```python
    def buildTree(self, preorder: List[int], inorder: List[int]) -> TreeNode:
        id_map = {val:idx for idx,val in enumerate(inorder)}
        
        def helper(in_left, in_right):
            if not preorder or in_left > in_right:
                return None
            
            val = preorder.pop(0)
            root = TreeNode(val)
            idx = id_map[val]
            
            root.left = helper(in_left, idx-1)
            root.right = helper(idx+1, in_right)
            return root
        
        return helper(0, len(inorder)-1)
```

### Iterative Solution

借助栈的先进后出，与 中序列表 从前向后 匹配，匹配成功的将其指定为 parent!

```python
def buildTree(self, preorder: List[int], inorder: List[int]) -> TreeNode:
    if not preorder or not inorder:
        return None

    root = TreeNode(preorder[0])
    stack = [root]
    inorder_index = 0

    for v in preorder[1:]:
        node = TreeNode(v)
        parent = None

        while stack and stack[-1].val == inorder[inorder_index]:
            inorder_index += 1
            parent = stack.pop()

        if parent:
            parent.right = node
        else:
            stack[-1].left = node

        stack.append(node)

    return root
```

preorder 和 inorder 都使用索引！

```python
def buildTree(self, preorder: List[int], inorder: List[int]) -> TreeNode:
    if not preorder:
        return None

    root = TreeNode(preorder[0])
    stack = [root]

    pre = 1
    ino = 0

    while pre < len(preorder):
        curr = TreeNode(preorder[pre])
        pre += 1
        prev = None

        while stack and stack[-1].val == inorder[ino]:
            prev = stack.pop()
            ino += 1

        if prev:
            prev.right = curr
        else:
            stack[-1].left = curr

        stack.append(curr)

    return root
```

## 106 Inorder and Postorder

### Recursive Solution

后序后找跟，中序分左右<br>
后序直接传，中序索引割

#### 切片，递归

```python
    def buildTree(self, inorder: List[int], postorder: List[int]) -> TreeNode:
        if not inorder or not postorder:
            return None
        
        current = postorder.pop()
        node = TreeNode(current)
        index = inorder.index(current)
        node.right = self.buildTree(inorder[index+1:], postorder)
        node.left = self.buildTree(inorder[:index], postorder)
        
        return node
```

#### 字典，递归

```python
    def buildTree(self, inorder: List[int], postorder: List[int]) -> TreeNode:
        id_map = {val:idx for idx,val in enumerate(inorder)}
        
        def helper(in_start, in_end):
            if in_start > in_end or not postorder:
                return None
            
            val = postorder.pop()
            root = TreeNode(val)
            idx = id_map[val]
            
            root.right = helper(idx+1, in_end)
            root.left = helper(in_start, idx-1)
            return root
        
        return helper(0, len(inorder)-1)
            
```

### Iterative Solution

```python
    def buildTree(self, inorder: List[int], postorder: List[int]) -> TreeNode:
        
        if not inorder or not postorder:
            return None
        
        root = TreeNode(postorder[-1])
        stack = [root]
        inorder_index = len(inorder)-1
        postorder_index = inorder_index-1
        
        while postorder_index >= 0:
            v = postorder[postorder_index]
            node = TreeNode(v)
            parent = None
            postorder_index -= 1
            while stack and stack[-1].val == inorder[inorder_index]:
                inorder_index -= 1
                parent = stack.pop()
            
            if parent:
                parent.left = node
            else:
                stack[-1].right = node
                
            stack.append(node)
            
        return root
```

## 889 Preorder and Postorder

[889. Construct Binary Tree from Preorder and Postorder Traversal](https://leetcode.com/problems/construct-binary-tree-from-preorder-and-postorder-traversal/)

### Recursive Solution

```python
    def constructFromPrePost(self, pre: List[int], post: List[int]) -> TreeNode:
        self.pre_index = 0
        self.post_index = 0
        def dfs(pre, post):
            root = TreeNode(pre[self.pre_index])
            self.pre_index += 1
            if root.val != post[self.post_index]:
                root.left = dfs(pre, post)
            if root.val != post[self.post_index]:
                root.right = dfs(pre, post)
            self.post_index += 1
            return root
        return dfs(pre, post)
```

### Iterative Solution

```python
    def constructFromPrePost(self, pre: List[int], post: List[int]) -> TreeNode:
        stack = [TreeNode(pre[0])]
        post_index = 0
        
        for v in pre[1:]:
            node = TreeNode(v)
            while stack[-1].val == post[post_index]:
                stack.pop()
                post_index += 1
            if not stack[-1].left:
                stack[-1].left = node
            else:
                stack[-1].right = node
            stack.append(node)
        return stack[0]
```
