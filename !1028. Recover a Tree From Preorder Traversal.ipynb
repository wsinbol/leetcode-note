# [1028. Recover a Tree From Preorder Traversal](https://leetcode.com/problems/recover-a-tree-from-preorder-traversal/)

We run a preorder depth first search on the root of a binary tree.

At each node in this traversal, we output D dashes (where D is the depth of this node), then we output the value of this node.  (If the depth of a node is D, the depth of its immediate child is D+1.  The depth of the root node is 0.)

If a node has only one child, that child is guaranteed to be the left child.

Given the output S of this traversal, recover the tree and return its root.

## 思路版

通过题目的分析，可以得出层级和节点之间的关系。但是有一个问题，就是把两位数以上的数字给拆开了。。。所以在字符串处理的过程中要考虑多种情况，像这种情况要采取字符串连接形式拼装起来，而不是单单遍历字符串！

```
i = 0
level = defaultdict(list)
d = {}
for c in S:
    if c == '-':
        i += 1
    else:
        level[i].append(c)
        d[c] = i
        i = 0

print(level)
print(d)
```

```
    def recoverFromPreorder(self, S: str) -> TreeNode:
        stack,i = [],0
        while i < len(S):
            val,level = '',0
            while i < len(S) and S[i] == '-':
                level += 1
                i += 1
                
            while i < len(S) and S[i] != '-':
                val = val + S[i] # 这里要注意
                i += 1
                
            while len(stack) > level:
                stack.pop()
                
            node = TreeNode(val)
            if stack and stack[-1].left is None:
                stack[-1].left = node
            elif stack:
                stack[-1].right = node
                
            stack.append(node)
            
        return stack[0]
```

## DFS 版

解法非常巧妙，首先通过正则的方式将 值和层级 分开并一一对应起来，然后递归左子树，左子树递归完毕递归右子树。

```
def recoverFromPreorder(self, S: str) -> TreeNode:
    vals = [(len(s[1]), int(s[2])) for s in re.findall("((-*)(\d+))", S)]

    def fn(level):
        if not vals or level != vals[0][0]: 
            return None
        node = TreeNode(vals.pop(0)[1])
        node.left = fn(level+1)
        node.right = fn(level+1)
        return node
    return fn(0)
```
