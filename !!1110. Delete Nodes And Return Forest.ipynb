# [1110. Delete Nodes And Return Forest](https://leetcode.com/problems/delete-nodes-and-return-forest/)

Given the root of a binary tree, each node in the tree has a distinct value.

After deleting all nodes with a value in to_delete, we are left with a forest (a disjoint union of trees).

Return the roots of the trees in the remaining forest.  You may return the result in any order.

## 开篇总结

- Solve tree problem with recursion first.
- To remove a node, the child need to notify its parent about the child's existance.
- To determine whether a node is a root node in the final forest, we need to know [1] whether the node is removed (which is trivial), and [2] whether its parent is removed (which requires the parent to notify the child)

> It is very obvious that a tree problem is likely to be solved by recursion. The two components above are actually examining interviewees' understanding to the two key points of recursion:

- passing info downwards -- by arguments
- passing info upwards -- by return value

## 自己版

层序遍历的方式遍历树节点，提交的过程就是一层一层打补丁。。。

```python
class Solution:
    def delNodes(self, root: TreeNode, to_delete: List[int]) -> List[TreeNode]:
        queue = [(root,None)]
        
        if root.val not in to_delete:
            res = [root]
        else:
            res = []
            
        while queue:
            size = len(queue)
            for i in range(size):
                current,parent = queue.pop(0)
                if current:
                    queue.append((current.left, current))
                    queue.append((current.right, current))
                    
                    if current.val in to_delete:
                        tmp = current.val
                        if current.left and current.left.val not in to_delete:
                            res.append(current.left)
                        if current.right and current.right.val not in to_delete:
                            res.append(current.right)
                        if parent:
                            if parent.left and parent.left.val == tmp:
                                parent.left = None
                            else:
                                parent.right = None
        return res
```

## BFS Solution

比我写的简洁多了。。。但是效率没有我的高

```python
class Solution:
    def delNodes(self, root: TreeNode, to_delete: List[int]) -> List[TreeNode]:
        queue = collections.deque([(root, False)])
        res = []
        
        while queue:
            current, has_parent = queue.popleft()
            if not has_parent and current.val not in to_delete:
                res.append(current)
                
            has_parent = not current.val in to_delete
                
            if current.left:
                queue.append((current.left, has_parent))
                if current.left.val in to_delete:
                    current.left = None
                    
            if current.right:
                queue.append((current.right, has_parent))
                if current.right.val in to_delete:
                    current.right = None
                    
        return res
```

## 大神版 - DFS version

https://leetcode.com/problems/delete-nodes-and-return-forest/discuss/328854/Python-Recursion-with-explanation-Question-seen-in-a-2016-interview

```python
class Solution:
    def delNodes(self, root: TreeNode, to_delete: List[int]) -> List[TreeNode]:
        res = []
        def helper(node, parent_exists):
            if not node:
                return None
            
            if node.val in to_delete:
                node.left = helper(node.left, False)
                node.right = helper(node.right, False)
                return None
            else:
                if not parent_exists:
                    res.append(node)
                node.left = helper(node.left, True)
                node.right = helper(node.right, True)
                return node
            
        helper(root, False)
        return res
```
