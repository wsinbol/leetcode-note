# [!109. Convert Sorted List to Binary Search Tree](https://leetcode.com/problems/convert-sorted-list-to-binary-search-tree/)

Given a singly linked list where elements are sorted in ascending order, convert it to a height balanced BST.

For this problem, a height-balanced binary tree is defined as a binary tree in which the depth of the two subtrees of every node never differ by more than 1.

## 自己版

> 给定列表中的中间元素将会作为二叉搜索树的根，该点左侧的所有元素递归的去构造左子树，同理右侧的元素构造右子树。这必然能够保证最后构造出的二叉搜索树是平衡的。

可以想到二分法，但是写代码的过程中，还是忽略了几点：

1. head_node == tail_node 时，要 return None，完全没写这两行代码。暴露的问题是对递归的 base case 理解不透彻！
2. tree_node.right = bst(slow.next, tail_node)，没有给 slow 更新为 slow.next！！！暴露的问题是对链表的操作能力不足！！！

```
    def sortedListToBST(self, head: ListNode) -> TreeNode:
        def bst(head_node, tail_node):
            if head_node == tail_node:
                return 
            slow = head_node
            fast = head_node

            while fast != tail_node and fast.next != tail_node:
                fast = fast.next.next
                slow = slow.next

            tree_node = TreeNode(slow.val)
            tree_node.left = bst(head_node, slow)
            tree_node.right = bst(slow.next, tail_node)
            return tree_node
        
        r = bst(head, None)
        return r
```

## 中序遍历

> 中序遍历一棵二叉搜索树的结果是得到一个升序序列。

这个方法不容易想到，而且代码也不看太明白，还要细细琢磨！https://leetcode-cn.com/problems/convert-sorted-list-to-binary-search-tree/solution/you-xu-lian-biao-zhuan-huan-er-cha-sou-suo-shu-by-/

```
class Solution:

    def findSize(self, head):
        ptr = head
        c = 0
        while ptr:
            ptr = ptr.next
            c += 1
        return c


    def sortedListToBST(self, head):
        """
        :type head: ListNode
        :rtype: TreeNode
        """

        # Get the size of the linked list first
        size = self.findSize(head)

        # Recursively form a BST out of linked list from l --> r
        def convert(l, r):
            nonlocal head

            # Invalid case
            if l > r:
                return None

            mid = (l + r) // 2

            # First step of simulated inorder traversal. Recursively form
            # the left half
            left = convert(l, mid - 1)

            # Once left half is traversed, process the current node
            node = TreeNode(head.val)   
            node.left = left

            # Maintain the invariance mentioned in the algorithm
            head = head.next

            # Recurse on the right hand side and form BST out of them
            node.right = convert(mid + 1, r)
            return node
        return convert(0, size - 1)
```
