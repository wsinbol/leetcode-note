#  N-sum problem

## 2-sum

> nums 中可能有多对儿元素之和都等于 target，请你的算法返回所有和为 target 的元素对儿，其中不能出现重复。

```python
def twoSum(self, nums: List[int], target: int) -> List[List[int]]:
    nums.sort()
    left, right = 0, len(nums)-1
    res = []
    while left < right:
        total = nums[left] + nums[right]
        if total == target:
            res.append([nums[left], nums[right] ])
            while left < right and nums[left] == nums[left+1]:
                left += 1
            while left < right and nums[right] == nums[right-1]:
                right -= 1
            left += 1
            right -= 1
        elif total < target:
            left += 1
        else:
            right -= 1
```

## 3-sum

```python
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        res = []
        nums.sort()
        for i in range(len(nums)-2):
            if i > 0 and nums[i] == nums[i-1]:
                continue
            left, right = i+1, len(nums)-1
            while left < right:
                total = nums[i] + nums[left] + nums[right]
                if total > 0:
                    right -= 1
                elif total < 0:
                    left += 1
                else:
                    res.append([nums[i], nums[left], nums[right]])
                    while left < right and nums[left] == nums[left+1]:
                        left += 1
                    while left < right and nums[right] == nums[right-1]:
                        right -= 1
                        
                    left += 1
                    right -= 1
        return res
```

## 4-sum

<font color=red>4-sum问题的解决方法是拆解成3-sum问题！</font>

```python
class Solution:
    def fourSum(self, nums: List[int], target: int) -> List[List[int]]:
        nums.sort()
        res = []
        for i in range(len(nums)-3):
            if i == 0 or nums[i] != nums[i-1]:
                three_sum = self.threeSum(nums[i+1:],target-nums[i])
                for item in three_sum:
                    res.append([nums[i]]+item)
        return res
    
    def threeSum(self, nums, target):
        r = []
        nums.sort()
        for i in range(len(nums)-2):
            if i > 0 and nums[i] == nums[i-1]:
                continue
            left, right = i+1, len(nums)-1
            while left < right:
                total = nums[i] + nums[left] + nums[right]
                if total == target:
                    r.append([nums[i],nums[left],nums[right]])
                    while left < right and nums[left] == nums[left+1]:
                        left += 1
                    while left < right and nums[right] == nums[right-1]:
                        right -= 1
                    left += 1
                    right -= 1
                elif total > target:
                    right -= 1
                else:
                    left += 1
        return r
```

## N-sum

```python
class Solution:
    def fourSum(self, nums: List[int], target: int) -> List[List[int]]:
        nums.sort()
        results = []
        self.nSum(nums, target,4,[],results)
        return results
    
    def nSum(self, nums, target, n, result, results):
        if len(nums) < n or n < 2:
            return 
        
        nums.sort()
        if n == 2:
            left, right = 0, len(nums)-1
            while left < right:
                total = nums[left] + nums[right]
                if total == target:
                    results.append(result + [nums[left], nums[right]])
                    while left < right and nums[left] == nums[left+1]:
                        left += 1
                    while left < right and nums[right] == nums[right-1]:
                        right -= 1
                    left += 1
                    right -= 1
                elif total < target:
                    left += 1
                else:
                    right -= 1
        else:
            for i in range(0, len(nums)-n+1):
                # 剪枝，大幅提高效率！
                if target < nums[i]*n or target > nums[-1]*n:
                    break
                if i == 0 or i > 0 and nums[i] != nums[i-1]:
                    self.nSum(nums[i+1:], target-nums[i], n-1, result+[nums[i]], results)
```
