# [919. Complete Binary Tree Inserter](https://leetcode.com/problems/complete-binary-tree-inserter/)

A complete binary tree is a binary tree in which every level, except possibly the last, is completely filled, and all nodes are as far left as possible.

Write a data structure CBTInserter that is initialized with a complete binary tree and supports the following operations:

- CBTInserter(TreeNode root) initializes the data structure on a given tree with head node root;
- CBTInserter.insert(int v) will insert a TreeNode into the tree with value node.val = v so that the tree remains complete, and returns the value of the parent of the inserted TreeNode;
- CBTInserter.get_root() will return the head node of the tree.

## 自己版

这道题的难点在于给一个完全二叉树添加新的节点，并且还要返回新的节点的父节点。先来看下 BFS 版本的：

```
class CBTInserter:

    def __init__(self, root: TreeNode):
        self._root = root

    def insert(self, v: int) -> int:
        new_node = TreeNode(v)
        queue = [self._root]
        while len(queue) > 0:
            size = len(queue)
            for i in range(size):
                current = queue.pop(0)
                if current.left is None:
                    current.left = new_node
                    return current.val
                if current.right is None:
                    current.right = new_node
                    return current.val
                
                queue.append(current.left)
                queue.append(current.right)
        

    def get_root(self) -> TreeNode:
        return self._root
```

BFS的时间复杂度在插入的时候相当高了，因为每次插入都进行层序遍历，而我们可以通过初始化的时候将节点的索引和节点的信息存储在字典里，这样添加的时候直接调用，并更新字典就可以了！来看下DFS版：

```

    def __init__(self, root: TreeNode):
        self._root = root
        self._max_pos = 0
        self._dict = {}
        def dfs(node,pos):
            if node:
                self._max_pos = max(pos, self._max_pos) # 这里注意一下，直接写成 self._max_pos = pos 是错误的！
                self._dict[pos] = node
                dfs(node.left,pos*2)
                dfs(node.right,pos*2+1)
        dfs(root, 1)
            

    def insert(self, v: int) -> int:
        
        pos = self._max_pos + 1
        index = int(pos / 2)
        parent = self._dict[index]
        
        new_node = TreeNode(v)
        if parent.left is None:
            parent.left = new_node
        else:
            parent.right = new_node
            
        self._dict[pos] = new_node
        self._max_pos = pos
        return parent.val

    def get_root(self) -> TreeNode:
        return self._root
```

<font color=red>
这里注意一下字典的 key-value 在 DFS 和 BFS 之处的不同：<br>
    self._max_pos = pos
    1. DFS 遍历的时候不是按照节点从左到右的顺序遍历，而是从上到下跳着遍历，这样直接将 pos 赋值是不正确的，而是要取最大值<br>
    2. BFS 遍历的时候是按照顺序遍历的，每次得到的 pos 都比前一次大，所以可以直接使用上面的赋值代码。
