# [653. Two Sum IV - Input is a BST](https://leetcode.com/problems/two-sum-iv-input-is-a-bst/)

```
Given a Binary Search Tree and a target number, return true if there exist two elements in the BST such that their sum is equal to the given target.
```

## 自己版

```python
    def dfs(node, k):
        if node:
            if node.val == k:
                return True
            elif node.val < k:
                return dfs(node.right, k-node.val)
            elif node.val > k:
                return dfs(node.left, node.val - k)

    return dfs(root, k) or self.findTarget(root.left if root.left else None, k) or self.findTarget(root.right if root.right else None, k)
```

## BFS+HashSet

```python
    def findTarget(self, root: TreeNode, k: int) -> bool:
        ret = set()
        queue = [root]
        while queue:
            size = len(queue)
            for i in range(size):
                current = queue.pop(0)
                
                if k - current.val in ret:
                    return True
                ret.add(current.val)
                if current.left:
                    queue.append(current.left)
                if current.right:
                    queue.append(current.right)
                    
        return False
```

## DFS+HashSet

```python
    def findTarget(self, root: TreeNode, k: int) -> bool:
        
        def find(node, k):
            if node == None:
                return False
            if k - node.val in res:
                return True
            
            res.add(node.val)
            return find(node.left, k) or find(node.right, k)
        
        res = set()
        return find(root, k)
```

## BST+Inorder

```python
    def findTarget(self, root: TreeNode, k: int) -> bool:
        
        def inorder(node):
            if node:
                inorder(node.left)
                res.append(node.val)
                inorder(node.right)

        res = []
        inorder(root)
        l,r = 0,len(res)-1
        while l < r:
            total = res[l] + res[r]
            if total == k:
                return True
            
            if total < k:
                l += 1
            else:
                r -= 1
        return False
```

## BST

双层DFS查找！

```python
    def findTarget(self, root: TreeNode, k: int) -> bool:
        
        def dfs(node, cur, k):
            if cur == None:
                return False
            return search(node, cur, k-cur.val) or dfs(node, cur.left, k) or dfs(node, cur.right, k)
        
        def search(node, cur, value):
            if node == None:
                return False
            return (node.val == value and cur != node) or (node.val < value and search(node.right, cur, value)) or (node.val > value and search(node.left, cur, value))
        
        return dfs(root, root, k)
```
