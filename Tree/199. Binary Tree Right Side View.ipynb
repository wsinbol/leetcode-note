# [199. Binary Tree Right Side View](https://leetcode.com/problems/binary-tree-right-side-view/)

Given a binary tree, imagine yourself standing on the right side of it, return the values of the nodes you can see ordered from top to bottom.

## 自己版

这道题和 662 题有些类似，目标是求每一层最右边的节点的值，从上到下顺序遍历自然就是结果了！

BFS思路：

```
    def rightSideView(self, root: TreeNode) -> List[int]:
        if not root:
            return 

        queue = [(root,0)]
        res = {}
        
        while len(queue) > 0:
            size = len(queue)
            for i in range(size):
                current,level = queue.pop(0)
                res[level] = current.val
                
                if current.left:
                    queue.append((current.left, level+1))
                    
                if current.right:
                    queue.append((current.right, level+1))
                    
        return res.values()
```

DFS思路：

```
    def rightSideView(self, root: TreeNode) -> List[int]:
        if not root:
            return 
        
        res = {}
        def dfs(node, level):
            if node:
                res[level] = node.val
                dfs(node.left,level+1)
                dfs(node.right,level+1)
                
        dfs(root,0)
        return res.values()
```

上述两个版本都可以跑通所有的测试用例，但是效率比较低。看下高票答案吧。。。

## 高票 BFS

<font color=red>
1. res使用 list 结构，而不是 dict 结构，相对来说节省了空间 <br>
2. res 的添加，通过 res 与 level 的关系来完成<br>
3. 注意是先添加右节点，然后是左节点！！！
</font>

```
    def rightSideView(self, root: TreeNode) -> List[int]:
        if not root:
            return 

        queue = [(root,0)]
        res = []
        
        while len(queue) > 0:
            size = len(queue)
            for i in range(size):
                current,level = queue.pop(0)
                
                if current.right:
                    queue.append((current.right, level+1))
                
                if current.left:
                    queue.append((current.left, level+1))
                    
                if level == len(res):
                    res.append(current.val)
                    
        return res
```
