# [979. Distribute Coins in Binary Tree](https://leetcode.com/problems/distribute-coins-in-binary-tree/)

Given the root of a binary tree with N nodes, each node in the tree has node.val coins, and there are N coins total.

In one move, we may choose two adjacent nodes and move one coin from one node to another.  (The move may be from parent to child, or from child to parent.)

Return the number of moves required to make every node have exactly one coin.

## DFS 版

> 定义 dfs(node) 为这个节点所在的子树中金币的过载量，也就是这个子树中金币的数量减去这个子树中节点的数量。接着，我们可以计算出这个节点与它的子节点之间需要移动金币的数量为 abs(dfs(node.left)) + abs(dfs(node.right))，这个节点金币的过载量为 node.val + dfs(node.left) + dfs(node.right) - 1

```
    def distributeCoins(self, root: TreeNode) -> int:
        self.res = 0
        def dfs(node):
            if not node:
                return 0
            left, right = dfs(node.left),dfs(node.right)
            self.res += abs(left) + abs(right)
            return node.val + left + right -1
        dfs(root)
        return self.res
```

## [通俗易懂版](https://leetcode.com/problems/distribute-coins-in-binary-tree/discuss/222000/A-better-illustration-of-recursive-method)

> Recursive method is indeed to find needed coins in a bottom-up way.<br>1.If the node is null, it require no coins.<br>2.If the node does't have enough coins to feed its children and itself (require n), it need to ask n coins from its parent. So count = count + n<br>3.If the node have extra coins to feed its children and itself (extra n), it need to pass n coins from its parent. So count = count + n

```
    def distributeCoins(self, root: TreeNode) -> int:
        self.count = 0
        def count_need(node):
            if not node:
                return 0
            
            child_need = count_need(node.left) + count_need(node.right)
            if child_need + 1 == node.val:
                return 0
            elif child_need + 1 < node.val:
                send = node.val - (child_need + 1)
                self.count += send
                return -send
            elif child_need +1 > node.val:
                take = child_need + 1 - node.val
                self.count += take
                return take
            
            return 0
        
        count_need(root)
        return self.count
```

## 总结

首先这个题目有个隐含的信息，就是每一个节点至少有一枚硬币，而这个节点到其子树的所有叶子节点的路径总和迭代 +1 的结果。

这到题的难点在于如何发现规律，某个节点金币的过载量为 node.val + dfs(node.left) + dfs(node.right) - 1, 这个很容易可以理解！那么该节点与它的子节点之间需要移动的金币的数量就更好理解了：因为定义的 dfs函数返回的就是过载的金币数量，最终的结果势必要将过载的硬币全部抹去，那么移动的步数自然而然就是 dfs(node.left) + dfs(node.right)
