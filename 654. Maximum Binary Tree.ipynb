# [654. Maximum Binary Tree](https://leetcode.com/problems/maximum-binary-tree/)

Given an integer array with no duplicates. A maximum tree building on this array is defined as follow:

    1.The root is the maximum number in the array.
    2.The left subtree is the maximum tree constructed from left part subarray divided by the maximum number.
    3.The right subtree is the maximum tree constructed from right part subarray divided by the maximum number.

Construct the maximum tree by the given array and output the root node of this tree.

## 自己参考版

看完题目也想到了用 DFS 方法，核心代码和正确答案是基本一样的，只是细节方面把握不足，所以参考了一下！

```
    def constructMaximumBinaryTree(self, nums: List[int]) -> TreeNode:
        
        '''
        def getMax(l,r):
            max_index = l
            for i in range(l,r):
                if nums[i] > nums[max_index]:
                    max_index = i
                    
            return max_index
        '''
        
        def create(l,r):
            if l == r:
                return
                
            max_num = max(nums[l:r])
            max_index = nums.index(max_num)
            
            # max_index = getMax(l,r)

            new_node = TreeNode(nums[max_index])
            new_node.left = create(l,max_index)
            new_node.right = create(max_index+1,r)

            return new_node
            
        r = create(0,len(nums))
        return r
```

<font color=red>
这里提出几个小的意见：<br>
    1. 涉及到左、右指针的递归，从外面传参的时候最好用(0,len(nums)),而不是数组（列表）的最小索引和最大索引，这样在给列表切片时也不需要对右边界做特殊处理了！否则为了取到右边界的值，你还要进行 +1 操作，搞麻烦了！<br>
    2.一定要注意 return 的返回！
</font>

## BFS 版

```
    def constructMaximumBinaryTree(self, nums: List[int]) -> TreeNode:
        stack = []
        for num in nums:
            cur = TreeNode(num)
            while stack and stack[-1].val < cur.val:
                cur.left = stack.pop()
            if stack:
                stack[-1].right = cur 
            stack.append(cur)
        return stack[0]
```

这个解法非常巧妙，借助队列的特性、数字间大小关系判断应该在左子树还是在右子树，先看下作者提供的关键点：

> 1.We scan numbers from left to right, build the tree one node by one step;<br>
2.We use a stack to keep some (not all) tree nodes and ensure a decreasing order;<br>
3.For each number, we keep pop the stack until empty or a bigger number; The bigger number (if exist, it will be still in stack) is current number's root, and the last popped number (if exist) is current number's left child (temporarily, this relationship may change in the future); Then we push current number into the stack.
